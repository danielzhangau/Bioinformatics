### Creating Alignment class objects
two restrictions to keep in mind when creating new Alignment objects:

- The sequences must be the same length to be aligned
- If you have gaps in a sequence (denoted with -), you need to flag the sequence as 'gappy'

```python
from sequence import *
s1 = Sequence('THISLINE', Protein_Alphabet)
s2 = Sequence('ISALIGNED', Protein_Alphabet)
aln = Alignment([s1, s2])
```

```python
s1 = Sequence('THISLINE-', Protein_Alphabet)
```

```python
s1 = Sequence('THISLINE-', Protein_Alphabet, gappy = True)
aln = Alignment([s1, s2]) # now the alignment works!
print(aln)
```
