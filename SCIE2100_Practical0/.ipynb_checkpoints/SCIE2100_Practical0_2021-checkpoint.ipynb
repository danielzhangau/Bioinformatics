### Programming lingo - let's start from the bottom and work our way up

An **<span style="color:blue">object</span>** is something that exists in a defined manner – a unique instance of a data structure. It is recognised by the computer and the computer knows all its features because they are described in a data structure. This concept is quite abstract but you need to understand that a computer only knows what you tell it and if you can’t describe an object – you can’t build code. Luckily, in Python, a lot of objects come predefined: integer, string, double, float, list, dictionary, set, tuple, etc. You should know what each of these objects is and how to use them. Each of these object types is actually a data structure and there are special operations you can do to use these objects. 

In most cases, you will want to store an object using a specific name so you can keep track of it and pass it around your code. When you assign an object a name, you create a **<span style="color:blue">variable</span>**.

`Variable1 = Object1`

Now we have named our objects creating variables, what do we do to use the variables? We write a function. A **<span style="color:blue">function</span>** is a list of instructions (code) that we have defined and given a name. You don’t have to organise your code into functions but good programmers always do. A function can take input in the form of variables, execute code, and then return an answer. A function is flexible and can also work in a number of different ways: input -> process -> output is just standard. Below is an example of a function.

```python
def revString(string):  
    print(string)
    nString = "" #create an empty string  
    for c in string[::-1]: #iterate backward over the string  
        nString += c #add each character to the new string  
    print(nString)
```

Once we have loaded this function (selecting *Run* in Spyder), we can use it in the following way:

```python    
>>> revString('abcdefg')
abcdefg
gfedcba
```

You should now work through the following programming exercises, with the help of the tutor if necessary, to further practice Python programming. We are first going to implement basic operations on biological sequence data (specifically DNA, RNA and proteins).

**<span style="color:darkred">If at any point throughout these exercises you are not sure where to start, or you're already feeling a bit lost, we suggest that you review the Python Guide which covers all the basics mentioned above in more detail. introduction.py and bonus_code.py also contain a number of examples that you can run using Spyder.</span>**

------------------------------------------------------------------------------------------------------------------------------
   
**Exercise 1 : Create a simple counting function** 

Create a string using the DNA sequence *exercise1_seq* below, which can also be found in *introduction.py*. Use the `string`’s [`count`] (https://www.tutorialspoint.com/python/string_count.htm) function to count the number of adenine (A), cytosine (C), guanine (G) and thymine (T) bases in the string. You might consider a `for` loop to iterate over each character that must be counted. Your function should print the counts out like so:

```

```

------------------------------------------------------------------------------------------------------------------------------
**<SPAN STYLE="color:#5F018B;font-size:18pt;weight:bold">Part 2: More complex Python structures</SPAN>**

### After completing Part 2 you should be familiar with:
* The structure of a class and how to access its functionality
* The purposes and structures of constructors, class methods, and class variables
* How to edit and use classes

### Working with Classes

Imagine you are working with sequence data, for example DNA. DNA could be represented as a string object but you want to store lots of sequences and you want to associate each sequence with specific features like an alphabet. You also want to be able to perform special operations specifically on your sequences. If you stick with the string data structure – you are limited in what you can do to your sequence. What you really want is a new data structure that will describe a sequence along with all of its special features – this is called a *class*.

Open `guide.py`. What is in this file? 

Let’s review some of the terminology that is used in the guide.py module. The term `def` is used to describe a function or a class method. *A function is defined outside a class while a class method is defined within a class.* Using class methods is more complex than other functions as we must first create an object based on a class, called an instance. 

Indentation is key in Python and that is one way you can distinguish class methods and functions. The first line of a class method (containing the term `def`) will be indented underneath a line that starts with the term `class`, while the first line of a function (containing the term `def`) will not be indented. 

Another difference between functions and class methods is that class methods always use the term `self`. For example, you could have lots of `Sequence` instances, so `self` is an argument that tells the computer which `Sequence` instance to use. You **do not** specify what `self` is when you create a class instance or call a class method; it is built into the class and class methods. *The Python Guide introduces classes with additional detail and examples if you would like more information.*

The `guide.py` module contains various classes, each with a different set of analysis tools/approaches for sequences and alignments. Consider the below questions:

* How many *class methods* do each of the *seven classes* in the `guide.py` module have?

* How many *functions* are in the `guide.py` module

------------------------------------------------------------------------------------------------------------------------------
**Exercise 5 : Introducing the Alphabet Class**	

To work with the classes in the following exercises, you will need to import `guide.py`. If you don’t import the module you are interested in using, Python will not be able to recognise any of the functions or classes you are trying to use from the module. Run the following in the interpreter to import `guide.py`:

```python
from guide import *
```

### FASTA file format

Biological sequences could just be supplied as the sequence by itself in a text file, but you can see why this would quickly get confusing when dealing with the massive amounts of sequence data that is available. Many file formats have been developed to store sequence data to help organisation. The most common format, and the one we will use is the FASTA format. A FASTA file can store one or more nucleotide or protein sequences, and looks like this:

```
>SequenceA
AGCTCCGCGATATACCATAAAACGTA
>SequenceB, with some more information
ACGCTAGCTAGCTGCGCGCTATATATGCGCATAGATCGTCGCG
AGTCGCTCGTAGCTAGTAGTCG
```

Each sequence in the file begins with a line starting with a > followed by the sequence’s name and possibly other unstructured information. The sequence itself starts on the following line, and ends when a line starts with a >, or the file ends.

FASTA files are simple and easy to read. This also makes it easy to write code to read and write them. We have done this for you, and in *guide.py* you can use the `readFastaFile` and `writeFastaFile` functions to read and write lists of sequences from and to FASTA files. Note that `readFastaFile` always returns a list, even if there is only one sequence in the FASTA file. You will need to select a specific sequence from the list to use it.
